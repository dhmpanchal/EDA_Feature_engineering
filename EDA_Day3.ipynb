## Flight Price Prediction (EDA + Feature Engineering)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
%matplotlib inline
```

```python
df_train=pd.read_excel('dataset/Data_Train.xlsx')
df_train.head()
```

```python
df_test=pd.read_excel('dataset/Test_set.xlsx')
df_test.head()
```

```python
#combine train and test data
df = pd.concat([df_train, df_test], axis=0)
df.tail()
```

```python
df.info()
```

### Feature Engineering Process

```python
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey', axis=1, inplace=True)
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
df.head()
```

```python
df['Arrival_Hour']=df['Arrival_Time'].apply(lambda x:x.split(':')[0])
df['Arrival_Min']=df['Arrival_Time'].apply(lambda x:x.split(':')[1])
df.head(1)
```

```python
df['Arrival_Hour'] = df['Arrival_Hour'].astype(int)
df['Arrival_Min'] = df['Arrival_Min'].astype(int)
```

```python
df.drop('Arrival_Time', axis=1, inplace=True)
df.head(1)
```

```python
df['Dep_Hour']=df['Dep_Time'].apply(lambda x:x.split(':')[0])
df['Dep_Min']=df['Dep_Time'].apply(lambda x:x.split(':')[1])
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Min'] = df['Dep_Min'].astype(int)
df.drop('Dep_Time', axis=1, inplace=True)
df.head(1)
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0, '1 stop':1, '2 stops':2, '3 stops':3, '4 stops':4, 'nan':1})
```

```python
df.drop('Route', axis=1, inplace=True)
```

```python
df.head()
```

```python
# converting duration feature into minuets
def convert_duration_to_minuets(duration):
    parts=duration.split()
    total_min = 0
    for part in parts:
        if part[-1] == 'h':
            total_min += int(part[:-1])*60
        elif part[-1] == 'm':
            total_min += int(part[:-1])
    
    return total_min
```

```python
df['Duration_min'] = df['Duration'].apply(convert_duration_to_minuets)
```

```python
df.drop('Duration_in_minutes', axis=1, inplace=True)
```

```python
df.head()
```

```python
df.drop(6474, axis=0, inplace=True)
df.drop(2660, axis=0, inplace=True)
```

```python
df.drop('Duration', axis=1, inplace=True)
```

### label encoding with categorical data

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
```

```python
df['Airline']=label_encoder.fit_transform(df['Airline'])
df['Source']=label_encoder.fit_transform(df['Source'])
df['Destination']=label_encoder.fit_transform(df['Destination'])
df['Additional_Info']=label_encoder.fit_transform(df['Additional_Info'])
```
